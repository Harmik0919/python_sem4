### 04) WAP to demonstrate the use of Pie chart.

```python
degree=["M.Tech","B.Tech","Diploma","BSC(H)","BSC(IT)"]
student=[200,400,150,50,45]
cx=["r","y","g","b","c"]
plt.pie(student,labels=degree,colors=cx,autopct="%1.2f%%",radius=0.8)
plt.title("Degree Wise student list")
plt.show()
```

### 05) WAP to demonstrate the use of Bar chart.

```python
x=["Jan","Feb","March","April","May","June"]
y=[20,50,60,40,70,30]
y1=[30,40,50,20,40,50]
plt.bar(x,y,width=0.5)
plt.show()

plt.barh(x,y)
plt.show()

wd=0.3
length=range(len(x))
plt.bar([ i-wd/2 for i in length],y,width=wd,label="y")
plt.bar([ i+wd/2 for i in length],y1,width=wd,label="y1")
plt.legend()
plt.xticks(length,x)
plt.show()
```

### 07) WAP to demonstrate the use of Histogram.

### 08) WAP to display the value of each bar in a bar chart using Matplotlib.

```python
x = ["Jan", "Feb", "March", "April", "May", "June"]
y = [20, 50, 60, 40, 70, 30]
bar = plt.bar(x, y, width=0.5)

for i in bar:
    yc = i.get_height()
    plt.text(i.get_x() + i.get_width()/2, yc + 1, f"{yc}", ha='center')

plt.show()
```
